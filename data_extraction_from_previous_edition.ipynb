# Manipulando o Arquivo Word com python-docx

```python
from docx import Document
```

```python
doc = Document("/content/drive/MyDrive/Dados/Catálogo de Fornecedores/cea_teste.docx")
```

```python
empresas_cea = pd.DataFrame(columns=['Nome da Empresa', 'Serviços Fornecidos', 'Telefone', 'E-mail', 'Site', 'Descrição'])
```

```python
empresas_indx = 0
for paragraph in doc.paragraphs:
  if 'Nome: ' in paragraph.text:
    empresas_indx += 1
    empresas_cea.loc[empresas_indx,'Descrição'] = ''
    empresas_cea.loc[empresas_indx,'Nome da Empresa'] = paragraph.text.split(': ')[-1]
  elif 'Telefone: ' in paragraph.text:
    empresas_cea.loc[empresas_indx,'Telefone'] = paragraph.text.split(': ')[-1]
  elif 'E-mail: ' in paragraph.text:
    empresas_cea.loc[empresas_indx,'E-mail'] = paragraph.text.split(': ')[-1]
  elif 'Site: ' in paragraph.text:
    empresas_cea.loc[empresas_indx,'Site'] = paragraph.text.split(': ')[-1]
  elif 'Serviço fornecido / Supplied service: ' in paragraph.text:
    empresas_cea.loc[empresas_indx,'Serviços Fornecidos'] = paragraph.text.split(': ')[-1]
  
  if 'Imagens' in paragraph.text:
    continue
  
  if empresas_indx >= 1:
    empresas_cea.loc[empresas_indx,'Descrição'] += (' '+paragraph.text)
```

```python
empresas_cea['Descrição'] = empresas_cea['Descrição'].apply(lambda x: x.split('Descrição do fornecedor / Supplier description: ')[-1])
```

```python
empresas_cea
```

```python
from ast import Tuple
def splittingDescriptionByLanguage(full_description: str) -> Tuple(str, str):
  from googletrans import Translator
  translation = Translator()

  if '.' not in full_description:
    return ('', '')

  
  periods = full_description.split('.')
  portuguese = []
  english = []

  for period in periods:
    language = translation.detect(period).lang
    if language == 'en':
      english.append(period)
    if language == 'pt':
      portuguese.append(period)

  return ('.'.join(portuguese) , '.'.join(english))
```

```python
empresas_cea['Descrição (Português)'] = empresas_cea['Descrição'].apply(lambda x: splittingDescriptionByLanguage(x)[0])
empresas_cea['Descrição (Inglês)'] = empresas_cea['Descrição'].apply(lambda x: splittingDescriptionByLanguage(x)[1])
```
